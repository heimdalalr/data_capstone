# Stocks vs Cryptocurrency stability 
### Project Summary

This project is to compare the stability and growth of stocks and cryptocurrency as an investment. The raw data of this project was sourced from yahoo finance as csv files. The scope of the project is the year of 2021.

### Data Dictionary
| date           | datetime | date of the data                                                      |
|----------------|:--------:|-----------------------------------------------------------------------|
| open           | float    | opening price of given time frame                                     |
| high           | float    | high price of given time frame                                        |
| low            | float    | low price of given time frame                                         |
| close          | float    | close price of given time frame                                       |
| adjusted close | float    | adjusted close price adjusted for splits, dividends and capital gains |
| volume         | integer  | number or stocks or cryptocurrency traded in a given time frame       |

## Stocks basic exploratory information

Apple (AAPL) and Microsoft were chosen for this comparison since they are the top two stocks based off of activity of trades. The initial exploration are the minimums, maximums and the closing average that the stock ends with on a daily basis.

### Apple AAPL intial information

```python
print('\nYearly low: ', aapl['Low'].min())
print('\nYearly high: ', aapl['High'].max())
print('\nDaily close average: ', aapl['Close'].mean())
```

```python
plt.style.use('ggplot')

aapl_lows = aapl['Low']
aapl_high = aapl['High']
aapl_close = aapl['Close']
aapl_date = aapl['Date']

plt.plot(aapl_date, aapl_lows, linewidth="1", label="lows")
plt.plot(aapl_date, aapl_high, linewidth="1", label="highs")
plt.plot(aapl_date, aapl_close, linewidth="1", label="close")

plt.title('Apple Highs, Lows and Closing Values')
plt.legend()
```

## Cleaning data

Dropping volume column, for ease of use. Since it won't be used in the plotting or filtering of information.

```python
aapl.fillna(value={'depth': 0}, inplace=True)
print('\nReplace nan volume values with 0:\n', aapl['Volume'])
print('\nDrop all rows with nan values (Not in place):\n', aapl.dropna())
print('\nDrop all rows with nan values in certain columns (Not in place):\n', aapl.dropna(subset=['Low']))
```
